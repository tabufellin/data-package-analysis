```python
# Importar las librerías necesarias
from scapy.all import rdpcap
import pandas as pd

# 1. Descarga el archivo analisis_paquetes.pcap y asígnalo a una variable
packets = rdpcap('analisis_paquetes.pcap')

# 2. Convierte la variable a un DataFrame
df = pd.DataFrame([(p.src, p.dst, p.sport, p.dport, p.len) for p in packets], columns=['Src Address', 'Dst Address', 'Src Port', 'Dst Port', 'Length'])

# 3. Muestra el contenido de las primeras 5 filas del dataset
print("Primeras 5 filas del dataset:")
print(df.head())

# 4. Muestra los valores de las columnas Src Address, Dst Address, Src Port y Dst Port
print("\nValores de las columnas Src Address, Dst Address, Src Port y Dst Port:")
print(df[['Src Address', 'Dst Address', 'Src Port', 'Dst Port']])

# 5. Estadísticas
# a. Muestra todas las IP origen
print("\nTodas las IP origen:")
print(df['Src Address'].unique())

# b. Muestra todas las IP destino
print("\nTodas las IP destino:")
print(df['Dst Address'].unique())

# c. Encuentra la IP origen más frecuente
most_frequent_src = df['Src Address'].value_counts().index[0]
print("\nIP origen más frecuente:", most_frequent_src)

# Filtra el DataFrame para obtener solo las filas con la IP origen más frecuente
frequent_src_df = df[df['Src Address'] == most_frequent_src]

# i. Encuentra la IP destino a la que se comunica con más frecuencia la IP origen más frecuente
most_frequent_dst = frequent_src_df['Dst Address'].value_counts().index[0]
print("IP destino a la que se comunica con más frecuencia:", most_frequent_dst)

# ii. Encuentra el puerto destino al que se comunica la IP origen más frecuente
most_frequent_dport = frequent_src_df['Dst Port'].value_counts().index[0]
print("Puerto destino al que se comunica con más frecuencia:", most_frequent_dport)

# Busca información sobre el pr
```

```python
#7

# a. Crear un nuevo DF que incluya únicamente las conexiones con la dirección IP origen más frecuente
frequent_src_df = df[df['Src Address'] == most_frequent_src]
frequent_src_df
```

```python
# b. Crear un nuevo DF que utilice el DF anterior con las columnas src, dst y payload, y agruparlas por dst y la suma del payload
grouped_df = frequent_src_df.groupby(['Src Address', 'Dst Address'])['Length'].sum().reset_index()
grouped_df
```

```python
# c. Obtener la IP destino que más ha intercambiado bytes con la IP más frecuente
suspicious_dst = grouped_df.loc[grouped_df['Length'].idxmax(), 'Dst Address']
print("IP destino sospechosa:", suspicious_dst)
```

```python
# d. Crear un nuevo DF con la conversación entre la IP más frecuente y la IP sospechosa
conversation_df = frequent_src_df[(frequent_src_df['Dst Address'] == suspicious_dst) | (frequent_src_df['Src Address'] == suspicious_dst)]
conversation_df
```

```python
# e. Obtener los payloads del DF del inciso anterior y añadirlos a un array
payloads = []
for _, row in conversation_df.iterrows():
    packet = [p for p in packets if p.src == row['Src Address'] and p.dst == row['Dst Address']][0]
    payloads.append(packet.load)

payloads
```

```python
# f. Mostrar el contenido del array de payloads
print("Contenido de los payloads:")
for payload in payloads:
    print(payload)
```

```python
# g. Observar los primeros bytes del contenido y describir si se encuentra algún dato que no tenga sentido enviar al puerto destino
print("Primeros bytes de los payloads:")
for payload in payloads:
    print(payload[:10])  # Mostrar los primeros 10 bytes de cada payload
```
